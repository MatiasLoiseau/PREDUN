```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('../data-private/CURSADA_HISTORICA_final.csv', parse_dates=['FECHA', 'FECHA_VIGENCIA'])
```

```python
data['ANIO'] = data['ANIO'].astype(int)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data['NOTA'], bins=10, kde=True)
plt.title('Distribution of NOTA')
plt.xlabel('NOTA')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='RESULTADO', data=data)
plt.title('Count of RESULTADO')
plt.xlabel('RESULTADO')
plt.ylabel('Count')
plt.show()
```

```python
avg_nota_by_resultado = data.groupby('RESULTADO')['NOTA'].mean().reset_index()
avg_nota_by_resultado
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='RESULTADO', y='NOTA', data=avg_nota_by_resultado)
plt.title('Average NOTA by RESULTADO')
plt.xlabel('RESULTADO')
plt.ylabel('Average NOTA')
plt.show()
```

```python
tipo_cursada_counts = data.groupby(['COD_CARRERA', 'TIPO_CURSADA']).size().unstack()
tipo_cursada_counts
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(tipo_cursada_counts, annot=True, fmt='g', cmap='viridis')
plt.title('TIPO_CURSADA Counts per COD_CARRERA')
plt.xlabel('TIPO_CURSADA')
plt.ylabel('COD_CARRERA')
plt.show()
```

```python
tipo_cursada_2_counts = data.groupby(['NOM_CARRERA', 'TIPO_CURSADA']).size().unstack()
tipo_cursada_2_counts
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(tipo_cursada_2_counts, annot=True, fmt='g', cmap='viridis')
plt.title('TIPO_CURSADA Counts per NOM_CARRERA')
plt.xlabel('TIPO_CURSADA')
plt.ylabel('NOM_CARRERA')
plt.show()
```

```python
records_per_year = data['ANIO'].value_counts().sort_index()
plt.figure(figsize=(10,6))
records_per_year.plot(kind='line', marker='o')
plt.title('Number of Records per Year')
plt.xlabel('Year')
plt.ylabel('Number of Records')
plt.show()
```

```python
# Get unique COD_CARRERA values
cod_carrera_list = data['COD_CARRERA'].unique()
num_carreras = len(cod_carrera_list)

# Determine the number of COD_CARRERA per chart
num_charts = 10
carreras_per_chart = int(np.ceil(num_carreras / num_charts))

# Plot the boxplots in separate charts
for i in range(num_charts):
    start_idx = i * carreras_per_chart
    end_idx = start_idx + carreras_per_chart
    subset_carreras = cod_carrera_list[start_idx:end_idx]
    subset_data = data[data['COD_CARRERA'].isin(subset_carreras)]

    if not subset_data.empty:
        plt.figure(figsize=(12,6))
        sns.boxplot(x='COD_CARRERA', y='NOTA', data=subset_data)
        plt.title(f'Distribution of NOTA by COD_CARRERA (Chart {i+1})')
        plt.xlabel('COD_CARRERA')
        plt.ylabel('NOTA')
        plt.show()
```

```python
pivot_table = data.pivot_table(values='NOTA', index='ANIO', columns='RESULTADO', aggfunc='mean')
pivot_table
```

```python
pivot_table.plot(figsize=(12,6))
plt.title('Average NOTA over Years by RESULTADO')
plt.xlabel('Year')
plt.ylabel('Average NOTA')
plt.legend(title='RESULTADO')
plt.show()
```
